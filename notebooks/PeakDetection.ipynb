# Peak Detection

The peak detection functionality can be used from peakutils library. Another
option is from Eli Billauer script: http://billauer.co.il/peakdet.html

```python
from matplotlib import pyplot as plt
from peakutils.plot import plot as pplot

import numpy as np
import pandas as pd
import peakutils
import matplotlib
import sys

%matplotlib inline
```

```python
# local
sys.path.insert(0, '../')
from pywim.utils import storage
```

```python
# open file test
f = storage.open_file('../data/wim_day_001_01_20170130.h5')
data = storage.dataset_to_dataframe(f[list(f.keys())[0]])
```

```python
data.plot()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
x = data.index.values

for k in data.keys():
    y = data[k].values
    indexes = peakutils.indexes(y, thres=0.5, min_dist=30)
    
    print(indexes)
    print(x[indexes], y[indexes])
    
    pplot(x, y, indexes)
plt.title('First estimate')
plt.grid(True)
plt.show()
```
